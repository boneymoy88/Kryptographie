```python
class Bruchzahlring(Ring):
    
    def __init__(self):
        self.null = 0
        self.eins = 1
        self._frier()
        
    def __str__(self):
        return "Q"
    
    def __eq__(self,other):
        if not type(self) == type(other):
            return False

        return True
    
    def element(self, info):
        return BruchzahlringElement(info, self)
    
    
    
class BruchzahlringElement(RingElement):
           
    def __init__(self, element, ring = Bruchzahlring()):
        
        # element ist entweder ein Tupel aus Zaehler und Nenner oder nur ein Integer (Nenner: 1)
        # ring kann zusaetzlich angegeben werden und soll den dazugehoerigen Bruchzahlring angeben
        
        # ueberpruefe ob tupel aus Integern (GanzzahlElementen) besteht
        if isinstance(element, tuple):
            zaehler, nenner = element[0], element[1]
            
            if not type(nenner) == int or nenner == 0:
                raise TypeError("Nenner ist keine Gazzahl oder 0")
            
            elif not type(zaehler) == int:
                raise TypeError("Zaehler ist keine Gazzahl")
            
            # falls Zaehler nicht null ist finde ggT von Zaehler und Nenner
            if not zaehler == 0:
                ggT = Ganzzahlring.ext_ggT(zaehler,nenner)[0]
                self.nenner = int(nenner/ggT)
                self.zaehler = int(zaehler/ggT)
            
            else:
                self.zaehler = ring.null
                
            self.ring = ring
            self.wert = zaehler/nenner
            
        # falls element als Integer angegeben wurde setze Nenner auf 1 
        elif isinstance(element, int):
            self.zaehler = element
            self.nenner = 1
            self.ring = ring
            self.wert = zaehler/nenner
            
        # falls element eine BruchzahlringElement ist speichere die Zahl selbst
        elif isinstance(element, BruchzahlringElement):
            self = element
        else: 
            raise RuntimeError("Das angegebene Objekt ist weder eine Ganzzahl noch eine Tupel oder ein BruchzahlElement")
        
        # Nenner soll nicht negativ sein
        if self.nenner < 0:
            self.nenner *= -1
            self.zaehler *= -1
    
    def drucke_element(self):
        return "Zaehler: [{0}] Nenner: [{1}] Wert: [{2}]".format(self.zaehler, self.nenner, self.wert)
    
    def __eq__(self,other):
        
        if not super().__eq__(other):
            return False
        return self.wert == other.wert
    
    def __neg__(self):
        return BruchzahlringElement((-self.zaehler, self.nenner), self.ring)
    
    def __add__(self,other):
        
        summand = BruchzahlringElement(other)
        
        # Berechnung des Ergebnisses
        return BruchzahlringElement((self.zaehler*summand.nenner + self.nenner*summand.zaehler, self.nenner*summand.nenner), self.ring)
    
    def __sub__(self,other):
        return -(self.__rsub__(other))
    
    def __rsub__(self,other):
        return other + (-self)
        
    def __rmul__(self,other):
        multiplikand = BruchzahlringElement(other)
        return BruchzahlringElement((self.zaehler * multiplikand.zaehler, self.nenenr * multiplikand.nenner), self.ring)
    
    def invers(self):
        h = self.zaehler
        self.zaehler = self.nenner
        self.nenner = h
        
        return BruchzahlringElement((self.zaehler, self.nenner), self.ring)
```
